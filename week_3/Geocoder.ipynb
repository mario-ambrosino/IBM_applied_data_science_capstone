# Geocoder

In this notebook, we will attach lat-long coordinates to the neighborhood / postal code dataset we obtained previously from Wikipedia.

Let's first import some libraries

```python
#!pip install geocoder
import requests
import pandas as pd
import numpy as np
```

Since the geocoder.google API was too slow in request, i used directly the dataset attached with the assigment.  

```python
# imports the Geo-Postal Code CSV
geo_url = "http://cocl.us/Geospatial_data"
r = requests.get(geo_url)
from io import StringIO
geo_pc = pd.read_csv(StringIO(r.text), index_col=0)

# generates the two dataframe to merge
pc_table = pd.read_csv("toronto.csv", index_col = 0)
geo_pc = pd.read_csv(StringIO(r.text), index_col=0)
```

Finally, I created a temporary array populated with the ordered sequence of (postal code, lat, long) values, then i transposed and attached as columns to the pc_table dataset.

```python
# generates the final datatable
final_pc = pc_table
temp_array = []
# populates a temp array with 
for ix in final_pc.index:
    temp_array.append([geo_pc[geo_pc.index == ix]["Latitude"][0],geo_pc[geo_pc.index == ix]["Longitude"][0]])
# transposes the list
latlon_list = np.array(temp_array).transpose()
# append two columns to the final dataset
final_pc["Latitude"]= latlon_list[0]
final_pc["Longitude"]= latlon_list[1]
```
